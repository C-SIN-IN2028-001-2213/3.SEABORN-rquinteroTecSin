# Gráficos de Distribución

En este JNB se revisará el tema de Gráficos de Distribución.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
iris_filepath = "https://raw.githubusercontent.com/C-SIN-IN2028-001-2213/3.SEABORN-rquinteroTecSin/main/data/iris.csv"

# Read the file into a variable iris_data
iris_data = pd.read_csv(iris_filepath, index_col="Id")

# Print the first 5 rows of the data
iris_data.head()
```

# Histogramas

Son el tipo de gráfico más comun para mostrar distribuciones.

```python
# Histogram 

sns.histplot(iris_data['Petal Length (cm)'])
```

# Gráfico de Densidad

Permiten mostrar las distribuciones de probabilidad de forma suavizada a diferencia de los histogramas que los muestran no suavizados.

```python
# KDE plot 

sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

```python
# KDE plot 

sns.kdeplot(data=iris_data['Sepal Width (cm)'], shade=True)
```

# Gráfico KDE bidimensional

Considera la representación suavizada de la distribución de datos para 2 variables (o columnas):

```python
# 2D KDE plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```

# Gráficos de distribución coloridos

Permiten distinguir distribuciones de probabilidad por categorías.

```python
# Add title
plt.title("Histogram of Petal Lengths, by Species")

# Histograms for each species
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')
```

```python
# Add title
plt.title("Distribution of Petal Lengths, by Species")

# KDE plots for each species
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', shade=True)
```
