# Gráficos de Dispersión

En este JNB se revisará el tema de Gráficos de Dispersión.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
insurance_filepath = "https://raw.githubusercontent.com/C-SIN-IN2028-001-2213/3.SEABORN-rquinteroTecSin/main/data/insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)

insurance_data
```

# Gráficos de Dispersión

Este tipo de gráfico muestra un conjunto de observaciones así como también un modelo de regresión que busca explicar el comportamiento de los datos.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

# Gráficos de dispersión coloridos

En este tipo de gráficos vamos a dividir la población en base a un factor adicional que permitiría poder detectar más de una correlación entre los datos y lo representamos con diversos colores.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

Si deseamos graficar la relación entre un dato categórico y otro continuo podemos utilizar el siguiente tipo de gráfico:

```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```
